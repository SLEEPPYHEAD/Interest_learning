```python
dataset.dropna(axis=0, how='any', inplace = True)
dataset = dataset.reset_index(drop=True)
```

### **PreProcessing:**

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

dataset['WindGustDir'] = encoder.fit_transform(dataset['WindGustDir'])
dataset['WindDir9am'] = encoder.fit_transform(dataset['WindDir9am'])
dataset['WindDir3pm'] = encoder.fit_transform(dataset['WindDir3pm'])
dataset['RainToday'] = encoder.fit_transform(dataset['RainToday'])
dataset['RainTomorrow'] = encoder.fit_transform(dataset['RainTomorrow'])
```

## **Dataset Spliting:**

```python
x=dataset.loc[:,dataset.columns!=target]
```

```python
x.head()
```

```python
y=dataset[target]
```

```python
y.head()
```

## **Dataset spliting in testing and training**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2)
```

## **Training Phase**

## **RandomForestt**

```python
%%time

from sklearn.ensemble import RandomForestClassifier   #importing libraries

clf = RandomForestClassifier()                        #creating method of algorithm 
                                                      #(Bracket a kichu na dile Default Parameter nibe)
                                                      #1st time model banaile default parameter newa valo
                                                      
clf.fit(x_train, y_train)                             #model fitting / data feeding / training

y_pred=clf.predict(x_test)                            #prediction
```

```python
from sklearn.metrics import accuracy_score

score=accuracy_score(y_test,y_pred)

print(score)
```

## **AdaBoost**

```python
from sklearn.ensemble import AdaBoostClassifier         #importing libraries


ada = AdaBoostClassifier()                              #creating method of algorithm 
                                                        #(Bracket a kichu na dile Default Parameter nibe)
                                                        #1st time model banaile default parameter newa valo

ada.fit(x_train, y_train)                               #model fitting / data feeding / training


y_pred = ada.predict(x_test)                                     #prediction



#clf.score(x, y)
```

## **GradientBoosting**

```python
from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier()
GB.fit(x_train, y_train)

y_pred = GB.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score

score=accuracy_score(y_test,y_pred)

print(score)
```

## **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(x_train, y_train)

y_pred =neigh.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score=accuracy_score(y_test,y_pred)

print(score)
```

## **Naive Byes**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train,y_train)

y_pred = gnb.predict(x_test)
```
